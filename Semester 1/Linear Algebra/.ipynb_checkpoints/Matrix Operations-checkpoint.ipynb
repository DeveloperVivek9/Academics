```python
def Solution():
    n=(int)(input("Enter the number of equations : "))
    v=(int)(input("Enter the number of variables : "))
    A=np.zeros((n,v))
    for i in range(0,n): 
        A[i]=input("Enter coefficients of the equations (where no coefficent = 0) : ").split()
    B=np.zeros((n,1))
    B=input("Enter the RHS of the equations in the same order as equation input : ").split()
    B = list(map(int, B)) #This is important since default datatype for input is string. Error was related to item size
    return (np.linalg.inv(A)).dot(B)
```

### Norm 

```python
#Where A -> Coordinate Values for n-norm
def Norm(A,n):
    sum=0
    for i in A:
        sum+=i**n
    return sum**(1/n)
```

```python
import math
```

```python
Norm([1,2],2)
```

```python
math.sqrt(5)
```

```python
5**(1/2)
```

### Unit vector 

```python
def Unit(A):
    sum=0
    for i in A:
        sum+=i**2
    return A/sum**(1/2)
```

```python
Unit(A)
```

```python
def angle(a,b):
    dot_product=sum([a*b for a,b in zip(a,b)])
    distances_product=Norm(a,2)*Norm(b,2)
    theta=math.acos(dot_product/distances_product)
    return round(math.degrees(theta),2)
```

```python
angle(a,b)
```
